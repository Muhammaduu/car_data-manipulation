# Q6) What is the average MSRP for cars based on their origin?

#### This questions need two to be carried out be answering it 
#### 1) removimg special characters such dolla sighn and comma
#### 2) converting to number columns 
##### remember MSRP is  object data type


```python
df.head(2)
```

```python
# Removing any extra commas or spaces if needed
df['MSRP'] = df['MSRP'].replace(',', '')

# Converting the column to float
df['MSRP'] = df['MSRP'].astype(float)

# Now you can apply numerical functions
mean_msrp = df['MSRP'].mean()

print("Mean:", mean_msrp)
```

```python
# Group by 'Origin' and calculate the mean of 'MSRP' 
avg_prices = df.groupby('Origin')['MSRP'].mean()

print(avg_prices)
```
